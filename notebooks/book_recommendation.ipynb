### Popularity based

This system is based on most popular books which are highly rated among them

```python
#For that we can consider the books which have been rated by more then 200 users with their highest popular rating 

ratings_with_name = ratings.merge(books, on="ISBN")
```

```python
ratings_with_name.head()
```

```python
ratings_with_name.shape
```

```python
#Book with more then 200 ratings

num_rating_df = ratings_with_name.groupby("book").count()[["ISBN"]]
```

```python
num_rating_df.rename(columns={
    "ISBN" : "no_ratings"
}, inplace = True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('book')["rating"].mean()
```

```python
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on = "book")
```

```python
popular_df
```

Now for the popularity recommendation system, we select the books which has at least got more then 200 ratings, so that it can be unbias to judge them, we can select top 20 books based on that.

```python
popular_df = popular_df[popular_df["no_ratings"]>200].sort_values("rating", ascending= False).head(20)
```

```python
popular_df.merge
```

```python
popular_df = popular_df.merge(books, on = "book").drop_duplicates("book")[["book", "author", "year", "image_link"]]
```

### Collaborative Filtering Based recommendation System

We will consider the two factor to consider this recommendation system.

User has at least rated more then 200 books. --> So that it is an experience user.

At least 50 time that particular book has been rated --> To get similarity between two users

```python
x = ratings_with_name.groupby("User-ID").count()["book"] > 200
```

```python
knowledge_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name["User-ID"].isin(knowledge_users)]
```

```python
y = filtered_rating.groupby("book").count()["rating"] >=50
```

```python
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating["book"].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index="book", columns="User-ID", values = "rating")
```

```python
pt.fillna(0, inplace = True)
```

```python
pt.shape
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(pt)
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(algorithm="brute")
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(pt.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

```python
distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    #Index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse = True) [1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend("Message in a Bottle")
```
